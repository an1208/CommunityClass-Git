```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
mouse_data = pd.DataFrame({
    'weight': [0.9, 1.8, 2.4, 3.5, 3.9, 4.4, 5.1, 5.6, 6.3],
    'size': [1.4, 2.6, 1.0, 3.7, 5.5, 3.2, 3.0, 4.9, 6.3]
})
```

```python
X = mouse_data[['weight']]
Y = mouse_data['size']
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
x_line = np.linspace(X.min(), X.max() , 100)
y_line = model.predict(x_line)
```

```python
plt.scatter(X,Y, color = 'black' , label = 'Data Points')
plt.plot(x_line, y_line , color = 'blue' , linewidth = 2, label = 'Regression line')

plt.title('Mouse Size vs Weight')
plt.xlabel('Weight')
plt.ylabel('Size')
plt.legend()
plt.grid(True)
plt.show()
```
